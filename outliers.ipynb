```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('/content/loan.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
q1= dataset["CoapplicantIncome"].quantile(0.25)
```

```python
q3= dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IRQ = q3-q1
```

```python
min_range = q1-1.5*IRQ
max_range = q3+1.5*IRQ
```

```python
min_range,max_range
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=dataset,x='CoapplicantIncome')
plt.show()
```

```python
dataset_new = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=dataset_new,x='CoapplicantIncome')
plt.show()
```

```python
from scipy.stats import zscore
```

```python
# Calculate z-scores for the 'CoapplicantIncome' column
z_scores = zscore(dataset['CoapplicantIncome'])

# Add the z-score as a new column (optional)
dataset['z_score'] = z_scores

# Keep only rows where z-score is between -3 and 3
df_filtered = dataset[(z_scores > -3) & (z_scores < 3)]
```
