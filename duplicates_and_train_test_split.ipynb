***ILOC AND LOC***

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

print(df)


# Get the first row
print(df.iloc[0])

# Get first two rows
print(df.iloc[0:2])

# Get value in first row, second column
print(df.iloc[0, 1])  # Output: 25
```

```python
# Get row with index label 0
print(df.loc[0])

# Get "Name" column
print(df.loc[:, 'Name'])

# Get value in row 0, column 'Age'
print(df.loc[0, 'Age'])  # Output: 25
```
