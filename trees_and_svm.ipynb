```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/content/age_salary_placed.csv')
#dataset = pd.read_csv('/content/age_salary_placed.csv')
```

***SUPPORT VECTOR MACHINE.(regression)***

```python
dataset3 = pd.read_csv('/content/cgpa_salary_dataset.csv')
```

```python
dataset3
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = dataset3[['CGPA']]   # or appropriate column
y = dataset3['Salary']
```

```python
sns.scatterplot(x="CGPA",y="Salary", data=dataset3)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR(kernel="linear")
svr.fit(x_train ,y_train)
print(svr.score(x_train, y_train))
print(svr.score(x_test, y_test))
```

```python
sns.scatterplot(x="CGPA",y="Salary", data=dataset3)
plt.plot(dataset3["CGPA"],svr.predict(x), color="red")
plt.show()
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Feature and target
X = dataset3[["CGPA"]]
y = dataset3["Salary"]

# Scaling
sc_X = StandardScaler()
sc_y = StandardScaler()
X_scaled = sc_X.fit_transform(X)
y_scaled = sc_y.fit_transform(y.values.reshape(-1, 1)).ravel()  # flatten for SVR

# Fit SVR
svr = SVR(kernel="linear")
svr.fit(X_scaled, y_scaled)
print(svr.score(x_train, y_train))
print(svr.score(x_test, y_test))
# Predict and inverse transform for proper plotting
y_pred_scaled = svr.predict(X_scaled)
y_pred = sc_y.inverse_transform(y_pred_scaled.reshape(-1, 1))

# Plot
sns.scatterplot(x="CGPA", y="Salary", data=dataset3, label="Actual")
plt.plot(X, y_pred, color="red", label="SVR Prediction")
plt.xlabel("CGPA")
plt.ylabel("Salary")
plt.title("SVR with Linear Kernel (Scaled Data)")
plt.legend()
plt.grid(True)
plt.show()
```
